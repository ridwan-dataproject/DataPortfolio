#### Menarik hanya data Nomor Handphone dari Kolom 'Catatan'  


```python
def number(row):
    # split row in 'Catatan Column'
    split_function = row['Catatan'].split()
    # assign as 'No number exists' if no 'DL:' at first
    if split_function[0] != 'DL:':
        return 'No Number Exists'
    else:
        # only grab the number only, index of 4
        split_function_number = split_function[4]
        # remove the comma from the last character
        remove_comma = split_function_number.replace(",","")
    
        return remove_comma
```

#### Hasil ekstrasksi  
**No Number Exists** berarti no handphone tidak dicantumkan dalam kolom **'Catatan'**  
**08xxxxxxxxx** berarti no handphone dicantumkan

```python
df.apply(lambda row: number(row), axis=1)
```

#### Buat kolom baru bernama **'phone_number'** ke dalam tabel

```python
df['phone_number'] = df.apply(lambda row: number(row), axis=1)
```

```python
df.head()
```

#### Kategorikan Tabel berdasarkan  kolom 'Nama Pelanggan' dan ' phone_number'

```python
df_name = df.groupby(['Nama Pelanggan','phone_number'], as_index=False)
```

```python
df_best =df_name[['Jumlah Produk']].agg('sum')
df_best
```
