```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# load the file
df = pd.read_excel('Sales_2019.xlsx')
```

```python
# show the dataframe of file
df.head()
```

```python
# show the 5-number summary
df.describe()
```

```python
#create histogram
df.hist('Total',bins=12)
```

```python
df.hist('Count_cabang',bins=50)
plt.xlabel('Jumlah Cabang per Order')
plt.ylabel('Count')
```

```python
# create linear regression between Count_cabang and Qty
sns.lmplot(x='Count_cabang',y='Qty',data=df)
```

```python
# create pairplot among variables
sns.pairplot(data=df)
```

```python
# create distribution plot of Count_cabang
sns.distplot(df['Count_cabang'],bins=50,color='purple')
```

```python
# create distribution plot of Total
sns.distplot(df['Total'],bins=100,color='purple')
```

```python
# create distribution plot of Qty
sns.distplot(df['Qty'],bins=50,color='purple')
```

```python
df.head()
```

```python
# add new column named qty_per_cabang
df['qty_per_cabang'] = df.Qty / df.Count_cabang
```

```python
# show the addition
df.head()
```

```python
# create linear regression between qty_per_cabang and qty
sns.lmplot(x='qty_per_cabang',y='Qty',data=df)
plt.xlabel('Rata-rata Order Per Cabang')
plt.ylabel('Jumlah Ayam')
```

```python
# create pairplot of these two
sns.pairplot(df, x_vars=["Count_cabang", "qty_per_cabang"], y_vars=["Qty"],
             height=5, aspect=.8, kind="reg")
```

```python
# Find the slope and intercept 
# between Count_cabang and Qty
from scipy import stats
```

```python
X = df['Count_cabang'].values
```

```python
Y = df['Qty'].values
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
print("slope: %f    intercept: %f" % (slope, intercept))
```

```python
# Find the slope and intercept 
# between qty_per_cabang and Qty
# Y is still the same as above
X1 = df['qty_per_cabang'].values
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X1, Y)
print("slope: %f    intercept: %f" % (slope, intercept))
```
