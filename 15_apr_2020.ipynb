```python
def label_owner (row):
       if row['nama_cabang'] == 'AGPG RESTO & CAFE' :
          return 'Internal'
       if row['nama_cabang'] == 'SRENGSENG' :
          return 'Internal'
       if row['nama_cabang'] == 'KEMANGGISAN' :
          return 'Internal'
       if row['nama_cabang'] == 'CIPONDOH MAKMUR':
          return 'Internal'
       if row['nama_cabang']  == 'GROGOL':
          return 'Internal'
       if row['nama_cabang'] == 'PURI':
          return 'Internal'
       if row['nama_cabang'] == 'SETIA BUDI 2':
          return 'Internal'
       if row['nama_cabang'] == 'TJ DUREN2':
        return 'Internal'
       return 'Eksternal'
```

```python
df_47.apply(lambda row: label_owner(row), axis=1)
```
